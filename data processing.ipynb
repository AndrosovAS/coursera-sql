```python
import pandas as pd
import numpy as np
```

```python
# Connect the required databases 
chicago_crime = pd.read_csv('https://data.cityofchicago.org/resource/ijzp-q8t2.csv')
cenus_data = pd.read_csv('https://data.cityofchicago.org/resource/kn9c-c2s2.csv')
schools = pd.read_csv('https://data.cityofchicago.org/resource/9xs2-f89t.csv')
```

```python
# Задание 1. Найдите общее количество преступлений, записанных в таблице Crime.
print('Total number of crimes: ', len(chicago_crime))
```

```python
# Задание 2. Получите первые 10 строк из таблицы CRIME.
chicago_crime.head(10)
```

```python
# Задание 3. По скольким преступлениям был произведен арест?
print('Arrests made: ', np.sum(chicago_crime['arrest']))
```

```python
# Задание 4. Какие уникальные виды преступлений совершались на автозаправочных станциях (GAS STATION)?
gas_station = chicago_crime.loc[chicago_crime.loc[:, 'location_description'] == 'GAS STATION']
gas_station['primary_type'].unique()
```

```python
# Задание 5. Получите из таблицы данных переписи CENUS_DATA все районы (community areas), названия которых начинаются с буквы «B».
community_area = chicago_crime['community_area'].unique()
criminal_areas = cenus_data.loc[cenus_data.loc[:, 'ca'].isin(community_area), 'community_area_name']
criminal_areas.loc[criminal_areas.str.startswith('B')]
```

```python
# Задание 6. Получите список школ в округах с 10 по 15, имеющих сертификат здоровья (healthy school certified)
schools.loc[(schools.loc[:, 'police_district'] >= 10) & 
            (schools.loc[:, 'police_district'] <= 15) & 
            (schools.loc[:, 'healthy_schools_certified_'] == 'Yes')]
```

```python
# Задание 7. Вычислите средний рейтинг безопасности (Safety Score) всех школ.
print('AVG safety_score: ', np.mean(schools['safety_score']))
```

```python
# Задание 8. Получите список из 5 районов с наибольшим средним количеством учеников, поступивших в колледж (College Enrollments).
college = schools.loc[schools.loc[:, 'name_of_school'].str.contains('College')]
college.sort_values(['college_enrollment_number_of_students_'], 
                    ascending=False)[['college_enrollment_number_of_students_', 'community_area_name']].head(5)
```

```python
# Задание 9. Найдите район с самым низким рейтингом безопасности (Safety Score), используя подзапрос.
schools.loc[schools.loc[:, 'safety_score'] == np.min(schools['safety_score'])]['community_area_name']
```

```python
# Задание 10. Получите информацию о доходе на душу населения в районе, где рейтинг безопасности школ равен 1.
schools_area = schools.loc[schools.loc[:, 'safety_score'] == 1]['community_area_name'].str.title()
pd.merge(cenus_data, schools_area, on='community_area_name')[['ca', 'community_area_name', 'per_capita_income_']]
```
